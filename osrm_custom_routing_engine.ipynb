```python
import requests
import json
import pandas as pd
import re
```

```python
tweaks = ["trunk=26,trunk_link=20,primary=26,primary_link=32,secondary=27,secondary_link=21,tertiary=19,tertiary_link=14,unclassified=17,residential=15,living_street=14",
"trunk=16,trunk_link=14,primary=17,primary_link=17,secondary=19,secondary_link=9,tertiary=15,tertiary_link=11,unclassified=15,residential=13,living_street=13",
"trunk=17,trunk_link=13,primary=17,primary_link=16,secondary=18,secondary_link=9,tertiary=16,tertiary_link=11,unclassified=15,residential=12,living_street=12",
"trunk=15,trunk_link=13,primary=15,primary_link=16,secondary=16,secondary_link=9,tertiary=14,tertiary_link=10,unclassified=13,residential=12,living_street=12"]
```

```python
tweaks[0]
```

```python
previous = 2
current = 3

for each in tweaks:
    # creating a variable and storing the text
    # that we want to search
    search_text = tweaks[previous]

    # creating a variable and storing the text
    # that we want to add
    replace_text = tweaks[current]

    # Opening our text file in read only
    # mode using the open() function
    with open(r'/home/lohith/Documents/git_push/GIS/routing_qa/osrm-backend-master/profiles/car.lua', 'r') as file:

        # Reading the content of the file
        # using the read() function and storing
        # them in a new variable
        data = file.read()

        # Searching and replacing the text
        # using the replace() function
        data = data.replace(search_text, replace_text)

    # Opening our text file in write only
    # mode to write the replaced content
    with open(r'/home/lohith/Documents/git_push/GIS/routing_qa/osrm-backend-master/profiles/car.lua', 'w') as file:

        # Writing the replaced data in our
        # text file
        file.write(data)

# Printing Text replaced
print("Text replaced")
```

```python
links = ["http://0.0.0.0:5000/route/v1/driving/-73.89726003181595,40.66777314;-73.972348,40.6288953",
"http://0.0.0.0:5000/route/v1/driving/-73.9072564,40.6685546;-73.9849372,40.6930517"]

distance = []
duration = []

for each in links:
    response = json.loads(requests.get(each).text)
    distance.append(response.get('routes')[0]['distance'])
    duration.append(response.get('routes')[0]['duration'])

for each in distance:
    print(str(each) + " KM")
```

```python
for each in duration:
    print(str(each) + " Mins")
```
